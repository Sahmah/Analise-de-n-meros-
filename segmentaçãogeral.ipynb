```python
#Importação e especificação do modelo importado
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
```

```python
#Passar a imagem que quer analisar
results = model ('yolo1.png')
```

```python
#Salvar os resultados
for result in results:
    boxes = result.boxes
    result.save(filename='result.png')
```

```python
#Mostrar a imagem do resultado
import matplotlib.pyplot as plt
import cv2

img = 'result.png'

image = cv2.imread(img) #Renderiza a imagem
plt.imshow(image) #Mostra a imagem
```
